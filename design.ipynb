## Terms

* Substance: A generic word for a sample, pool, index tag etc. All substances are also Locatable
* Container: A container can hold Locatables, i.e. Substances or other Containers.
* Locatable: Anything that can be put into a Container (both Containers and Substances are Locatable)
* Sample: A sample probably means what one would expect in a lab, an individual item that's being researched. Samples are instances of Substances.

## Actors

![Actors](diagrams/pngs/0-actors.png)

### Admin

Admins setup the system and have all the rights all other users have.

### Coordinator

A lab coordinator takes care of creating projects, contacts etc. They may have rights to break the general workflow.

### Research engineer (RE)

Research engineers generally work within a well defined workflow in order to make their job easier and limit chances of errors.
